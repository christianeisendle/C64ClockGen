```python
import numpy as np
```

```python
fin = 17.734475e6
fout = 7.875e6
```

```python
def calc_fout(fin, m, d, o):
    fout = fin * m/d/o
    fvco = fout * o
    print(f"fout: {fout}, fvco: {fvco}")
    return fout, fvco
```

```python
def find_divider(fin, fout):
    opt_settings = None
    error = fout
    for m in range(1, 4095):
        for d in range(1, 127):
            for o in range(0, 127):
                ftmp = fin * m/d/((o+2)*2)
                if np.abs(ftmp-fout) < error:
                    opt_settings = (m, d, o)
                    error = np.abs(ftmp-fout)
                    print(f"error: {error/fout*1e6:.2f}ppm, {opt_settings}")
    return opt_settings
```

```python
def calc_bandwidth(m, ipreg, rzreg, czreg):
    cz = 196e-12 + czreg * 217e-12
    cp = 15e-12
    Ip = (ipreg[0] + 2*ipreg[1]+4*ipreg[2]) * 6e-6
    rz = (rzreg[0] + 2* rzreg[1]+4* rzreg[2] + 8* rzreg[3])* 4e3
    cp_gain = Ip / (2*np.pi)
    vco_gain = 900e6*2*np.pi
    wc = (rz * cp_gain * vco_gain * cz) / (m*(cz+cp))
    fc = wc / (2*np.pi)
    print(f"bandwidth: {fc / 1e3:.2f}kHz")
    return fc

```

```python
def calc_phase_margin(fc, rzreg, czreg):
    cz = 196e-12 + czreg * 217e-12
    cp = 15e-12
    rz = (rzreg[0] + 2* rzreg[1]+4* rzreg[2] + 8* rzreg[3])* 4e3
    wz = 1 / (rz*cz)
    wc = 2*np.pi*fc
    wp = (cz+cp)/(rz*cp*cz)
    phase_margin = np.degrees(np.arctan(wc/wz)- np.arctan(wc/wp))
    print(f"phase margin: {phase_margin:.1f}°")
    return phase_margin
```
